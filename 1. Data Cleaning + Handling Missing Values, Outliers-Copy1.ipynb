# Why Airbnb

- Airbnb bila didefinisikan merupakan suatu layanan online penyedia jasa penyewaan properti berupa rumah sewa, apartemen dan tersebar di semua Negara di dunia. Perusahaan ini pun didirikan di San Francisco pada tahun 2008 dan merupakan pionir dari online marketplace untuk hospitality services. Analisis data yang disediakan oleh pihak Airbnb sangat penting untuk memberikan informasi yang krusial untuk perusahaan dan pengusaha airbnb.


# Why Airbnb Singapore
- Tidak diragukan lagi bahwa Negara Singapur merupakan salah satu Negara atau kota sebagai tujuan wisata dan perjalanan jangka pendek di Asia Tenggara. Singapur juga merupakan kota dengan sistem transportasi yang sangat efisien, dengan biaya hidup yang tinggi, tempat perbelanjaan yang menjadi idaman selain Eropa. 

# Problem 
- Para customer yang ingin mencari properti dan tempat untuk _travelling_ di singapore terkadang masih merasa kesulitan untuk memperkirakan seberapa layak harga tersebut untuk ditempatin. Fitur yang digunakan untuk mencari lokasi tempat penginapan juga masih tergolong tidak terlalu detail untuk memperkirakan harga properti yang ingin kita sewa. 


# Goals
- Dengan menggunakan model Price Prediction dengan metoda Supervised Machine Learning - Regression,  saya ingin mencoba membantu untuk menambahkan fitur untuk para pelanggan  melihat harga normal dari setiap properti yang disewakan. Jika harga properti memiliki gap yang besar antara prediksi dan harga aktual, pelanggan dapat melakukan pencarian lebih jauh lagi tentang properti tersebut, yang dapat dilihat dari fasilitas-fasilitas mewah yang disediakan oleh si penyewa dan berpengaruh terhadap kenaikan harga sewa di airbnb.


# Features Dictionary


id: Angka Unique untuk mengidentifikasi Airbnb

listing_url : Alamat url dari Airbnb yang mengarah kepada website Airbnb dimana listing tersebut disewakan

scrape_id : Angka unique yang didapatkan berdasarkan tanggal scrape

last_scrape : data tanggal terakhir Airbnb tersebut di ambil

name : Nama yang diberikan Host untuk listing Airbnb nya

summary : Ringkasan terkait informasi seputar listing Airbnb

space : Berisi tentang deskripsi space . exp: 3 kamar tidur, 1 kamar mandi

description : Berisi tentang deskripsi listing secara umum. exp: lokasi, jalan, neighbourhood

experiences_offered : Pengalaman yang ditawarkan Host kepada penyewa, biasanya kegiatan untuk mengeksplor event, kuliner, budaya, atau sekedar memperkenalkan tokyo .

neighborhood_overview : gambaran seputar area/neighbourhood lokasi listing tersebut berada. exp: area ini dapat diakses dengan 10 menit by train yang mana dekat dengan toko...

notes : catatan tertentu yang diberikan Host kepada pelanggan selama menginap di listing

transit : gambaran seputar transportasi umum dekat lokasi listing. exp : 10 menit ke Stasiun terdekat

access : deskripsi seputar ruangan/area yang dapat diakses di listing

interaction : informasi seputar host jika listing ingin berinteraksi. exp : saya tinggal tidak jauh dari anda di... apabila anda memerlukan pertolongan

house_rules : aturan seputar listing. biasanya berupa aturan pembuangan sampah, jam malam, kebisingan dan lain-lain

thumbnail_url : url berisi gambaran singkat listing.

picture_url : url berisi halaman foto-foto seputar airbnb.

host_id : Angka unique untuk mengidentifikasi host

host_url : Alamat url dari Host yang memberikan informasi lengkap seputar Host. exp : nama, jumlah listing, area, reviews

host_name : Nama dari Host

host_since : Tanggal awal menjadi host

host_location : lokasi host tersebut tinggal

host_about : berisi deskripsi seputar host. exp: nama saya ..., saya memiliki beberapa penginapan di ...

host_response_time : waktu rata-rata response sang host . exp: within a few hours

host_response_rate : rating penentuan waktu tersebut. exp : 100% berarti selalu merespon sesuai dengan ketentuan waktu yang ditetapkan

host_acceptance_rate : rating penerimaan untuk disewakan. exp : 60% berarti tidak selalu diterima atau cukup strict terhadap persyaratan

host_is_superhost : penilaian berdasarkan 4 tahun sekali terkait kinerja host berupa label t or f. berdasarkan pengalaman yang diberikan kepada pelanggan

host_listings_count : jumlah listing yang dimiliki host tersebut

host_verifications : berisi tentang media-media yang digunakan host untuk memverifikasi akun nya terhadap pihak Airbnb. exp: email, phone, government id

host_identity_verified : setelah mengikisi media-media yang digunakan host untuk memverifikasi akun nya. ini adalah verifikasi yang dilakukan host bahwa akun ini benar. exp: "Airbnb will ask you to confirm your identity to help us make sure you’re really you." terkadang banyak yang mengabaikan ini dengan alasan kesulitan berbahasa inggris

street : berisi data lokasi dengan spesifikasi neighbourhood, kota, dan negara. exp : Shibuya,Tokyo,Japan

neighbourhood : lokasi listing tersebut berada/ distrik. 

state : Negara. 

zipcode : kode pos

country_code : kode negara .

property_type : tipe properti dari listing. exp: House/ Apartment/ ...

room_type: tipe ruangan dari listing (Entire home/apt, Private room, or Shared room).

accommodates: jumlah orang yang listing dapat akomodasikan.

bathrooms: jumlah bathrooms

bedrooms: jumlah bedrooms

beds: jumlah tempat tidur

amenities : perlengkapan yang ditawarkan host untuk pengunjungnya. exp: Air Conditioning, Refrigerator, Wifi, ....

cleaning_fee : biaya yang ditentukan host untuk bersih-bersih.

security_deposit : biaya deposit yang ditentukan host untuk keamanan. Apabila ada terjadi sesuatu, uang tersebut digunakan untuk membayarnya. exp: setrika dirusak pengunjung, maka biaya ganti rugi ditanggung deposit ini. Apabila tidak ada apa-apa, uang akan dikembalikan saat selesai menginap.

minimum_nights : minimum hari

maximum_nights : maksimum hari menginap

availability_365 : jumlah ketersediaan listing per hari dalam setahun

number_of_reviews : jumlah reviews yang didapatkan host untuk listing tertentu (0-....) . 0 berarti tidak pernah disewa

review_scores_rating : nilai score dari review yang didapatkan (0-100). 0 berarti tidak pernah disewa

guests_included : jumlah perkiraan dapat menampung tamu. ada denda bila melebihi yang ditentukan

license : lisensi yang dimiliki host. exp : lisensi untuk melakukan pembuangan sampah pada perusahaan swasta , lisensi keamanan oleh pemadam kebakaran dll. Ada kesulitan di sini karena berbahasa Jepang

cancellation_policy : kebijakan pembatalan, biasanya ada maksimum hari untuk pembatalan apabila ingin refund. 7 hari sebelum tanggal menginap. bila lebih dari itu maka uang tidak bisa dikembalikan

reviews_per_month : jumlah review per bulan

price : harga listing

instant_bookable : apakah cara pemesanan property kepada hostnya dapat secara langsung atau membutuhkan persetujuan terlebih dahulu sebelum melakukan pemesanan

```python
import re
def remove_prefix(col):
    clean=re.sub("[$]", "", col) # fungsi untuk remove prefix  $ dengan ''.
    return clean
```

```python
def remove_comma(col):
    clean=re.sub(',','',col)
    return clean
```

```python
fixprice['security_deposit']=fixprice['security_deposit'].apply(remove_prefix)
fixprice['cleaning_fee']=fixprice['cleaning_fee'].apply(remove_prefix)
fixprice['security_deposit']=fixprice['security_deposit'].apply(remove_comma)
```

```python
fixprice['security_deposit']=fixprice['security_deposit'].astype(float) # mengubah tipe data yang sebelumnya object menjadi float
fixprice['cleaning_fee']=fixprice['cleaning_fee'].astype(float) # mengubah tipe data yang sebelumnya object menjadi float
```
